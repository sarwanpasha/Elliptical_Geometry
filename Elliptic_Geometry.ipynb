```python
# Manifolds with Constant Sectional Curvature
# Riemannian manifolds with constant curvature at every point are special cases of curved surfaces. They come in three forms, which implicitly define three geometries:

# Constant Positive Curvature: Elliptic geometry

# Constant Zero Curvature: Euclidean geometry

# Constant Negative Curvature: Hyperbolic geometry
```

```python
def elliptic_geometry(seq11=None, seq22=None):
    
    a=np.dot(seq11,seq22)
    b=np.linalg.norm(seq11)
    c=np.linalg.norm(seq22)
    
    small_value = 1e-8  # Small value to avoid division by very small numbers
    d = np.arccos(abs(a) / (b * c + small_value))
    
#     d=np.arccos(abs(a)/(b*c))
#     old haris
#     kernel_value = np.exp(-d)
#     new mine
    sigma = 1
    kernel_value = np.exp(-d**2 / (2 * sigma**2))
    return kernel_value

sequences=seq_data[:]

#%%
# Compute the kernel matrix
counter_check = 0
K = np.zeros((len(sequences), len(sequences)))
for i, seq1 in enumerate(sequences):
    print("i: ",i,"/",len(sequences))
    for j, seq2 in enumerate(sequences):
        if i <= j:
            K[i, j] = elliptic_geometry(seq11=seq1, seq22=seq2)
            K[j, i] = K[i, j]
```
